```python
number_rating = rating_with_books.groupby("title")["rating"].count().reset_index() 
```

```python
number_rating.rename(columns = {
    "rating" : "num of ratings"}, inplace=True)
```

```python
final_data_pair = rating_with_books.merge(number_rating, on="title")
final_data_pair.head(2)
```

```python
final_data_pair = final_data_pair[final_data_pair['num of ratings']>=50]
```

```python
final_data_pair.drop_duplicates(['user_id','title'])
```

```python
final_data_pair.shape
```

```python
books_pivot = final_data_pair.pivot_table(columns='user_id',index='title',values = 'rating')
books_pivot
```

```python
books_pivot.shape
```

```python
books_pivot.fillna(0 , inplace= True)
books_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(books_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors 
model = NearestNeighbors(algorithm='brute') 
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(books_pivot.iloc[237,:].values.reshape(1,-1))
```

```python
for i in range(len(suggestion)):
    print(books_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl','wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl','wb'))
pickle.dump(final_data_pair, open('artifacts/final_data_pair.pkl','wb'))
pickle.dump(books_pivot, open('artifacts/books_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(books_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(books_pivot.iloc[book_id,:].values.reshape(1,-1))

    for i in range(len(suggestion)):
        books = books_pivot.index[suggestion[i]]
        for j in books:
            print(j)
                       
```

```python
book_name = 'The Long Road Home'
recommend_book(book_name)
```
